```python
import torch
import matplotlib.pyplot as plt
```

```python
xs = torch.linspace(0, 10, 1000)
```

```python
ys = torch.sin(xs)
```

```python
xs.shape
```

```python
plt.plot(xs, ys)
```

```python
h = 0.01
```

```python
differential_matrix = torch.zeros((1000, 1000))
for i in range(0, 999):
  differential_matrix[i + 1][i] = -1
  differential_matrix[i][i + 1] = 1
```

```python
differential_matrix
```

```python
diff_ys = torch.mm(differential_matrix, ys.unsqueeze(1)) / (2 * h)
```

```python
cos_ys = torch.cos(xs)
```

```python
plt.plot(xs, ys)
plt.scatter(xs[1:-1], diff_ys.squeeze()[1:-1])
plt.plot(xs, cos_ys)
```
